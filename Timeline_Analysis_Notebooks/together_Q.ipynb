```python
def normalizing_funtion(df_1, df_2):
    normalized_df = df_1
    top_df = df_1
    bottom_df = df_2

    for i in range(len(bottom_df)):
    #     for each row (user)
        temp_list = []
        for j in (normalized_df.columns):
    #         for each column (month)
            top_total = top_df[j][i]
            bottom_total = bottom_df[j][i]
            if pd.isnull(bottom_total):
                temp_list.append(np.nan)
            elif top_total == 0 or bottom_total == 0:
                temp_list.append(0)
            else:
                if pd.isnull((top_total/bottom_total)):
                    temp_list.append(0)
                else:
                    temp_list.append((top_total/bottom_total))

        normalized_df['01-2009'][i] = temp_list[0]
        normalized_df['02-2009'][i] = temp_list[1]
        normalized_df['03-2009'][i] = temp_list[2]
        normalized_df['04-2009'][i] = temp_list[3]
        normalized_df['05-2009'][i] = temp_list[4]
        normalized_df['06-2009'][i] = temp_list[5]
        normalized_df['07-2009'][i] = temp_list[6]
        normalized_df['08-2009'][i] = temp_list[7]
        normalized_df['09-2009'][i] = temp_list[8]
        normalized_df['10-2009'][i] = temp_list[9]
        normalized_df['11-2009'][i] = temp_list[10]
        normalized_df['12-2009'][i] = temp_list[11]
        normalized_df['01-2010'][i] = temp_list[12]
        normalized_df['02-2010'][i] = temp_list[13]
        normalized_df['03-2010'][i] = temp_list[14]
        normalized_df['04-2010'][i] = temp_list[15]
        normalized_df['05-2010'][i] = temp_list[16]
        normalized_df['06-2010'][i] = temp_list[17]
        normalized_df['07-2010'][i] = temp_list[18]
        normalized_df['08-2010'][i] = temp_list[19]
        normalized_df['09-2010'][i] = temp_list[20]
        normalized_df['10-2010'][i] = temp_list[21]
        normalized_df['11-2010'][i] = temp_list[22]
        normalized_df['12-2010'][i] = temp_list[23]
        normalized_df['01-2011'][i] = temp_list[24]
        normalized_df['02-2011'][i] = temp_list[25]
        normalized_df['03-2011'][i] = temp_list[26]
        normalized_df['04-2011'][i] = temp_list[27]
        normalized_df['05-2011'][i] = temp_list[28]
        normalized_df['06-2011'][i] = temp_list[29]
        normalized_df['07-2011'][i] = temp_list[30]
        normalized_df['08-2011'][i] = temp_list[31]
        normalized_df['09-2011'][i] = temp_list[32]
        normalized_df['10-2011'][i] = temp_list[33]
        normalized_df['11-2011'][i] = temp_list[34]
        normalized_df['12-2011'][i] = temp_list[35]
        normalized_df['01-2012'][i] = temp_list[36]
        normalized_df['02-2012'][i] = temp_list[37]
        normalized_df['03-2012'][i] = temp_list[38]
        normalized_df['04-2012'][i] = temp_list[39]
        normalized_df['05-2012'][i] = temp_list[40]
        normalized_df['06-2012'][i] = temp_list[41]
        normalized_df['07-2012'][i] = temp_list[42]
        normalized_df['08-2012'][i] = temp_list[43]
        normalized_df['09-2012'][i] = temp_list[44]
        normalized_df['10-2012'][i] = temp_list[45]
        normalized_df['11-2012'][i] = temp_list[46]
        normalized_df['12-2012'][i] = temp_list[47]
        normalized_df['01-2013'][i] = temp_list[48]
        normalized_df['02-2013'][i] = temp_list[49]
        normalized_df['03-2013'][i] = temp_list[50]
        normalized_df['04-2013'][i] = temp_list[51]
        normalized_df['05-2013'][i] = temp_list[52]
        normalized_df['06-2013'][i] = temp_list[53]
        normalized_df['07-2013'][i] = temp_list[54]
        normalized_df['08-2013'][i] = temp_list[55]
        normalized_df['09-2013'][i] = temp_list[56]
        normalized_df['10-2013'][i] = temp_list[57]
        normalized_df['11-2013'][i] = temp_list[58]
        normalized_df['12-2013'][i] = temp_list[59]
        normalized_df['01-2014'][i] = temp_list[60]
        normalized_df['02-2014'][i] = temp_list[61]
        normalized_df['03-2014'][i] = temp_list[62]
        normalized_df['04-2014'][i] = temp_list[63]
        normalized_df['05-2014'][i] = temp_list[64]
        normalized_df['06-2014'][i] = temp_list[65]
        normalized_df['07-2014'][i] = temp_list[66]
        normalized_df['08-2014'][i] = temp_list[67]
        normalized_df['09-2014'][i] = temp_list[68]
        normalized_df['10-2014'][i] = temp_list[69]
        normalized_df['11-2014'][i] = temp_list[70]
        normalized_df['12-2014'][i] = temp_list[71]
        normalized_df['01-2015'][i] = temp_list[72]
        normalized_df['02-2015'][i] = temp_list[73]
        normalized_df['03-2015'][i] = temp_list[74]
        normalized_df['04-2015'][i] = temp_list[75]
        normalized_df['05-2015'][i] = temp_list[76]
        normalized_df['06-2015'][i] = temp_list[77]
        normalized_df['07-2015'][i] = temp_list[78]
        normalized_df['08-2015'][i] = temp_list[79]
        normalized_df['09-2015'][i] = temp_list[80]
        normalized_df['10-2015'][i] = temp_list[81]
        normalized_df['11-2015'][i] = temp_list[82]
        normalized_df['12-2015'][i] = temp_list[83]
        normalized_df['01-2016'][i] = temp_list[84]
        normalized_df['02-2016'][i] = temp_list[85]
        normalized_df['03-2016'][i] = temp_list[86]
        normalized_df['04-2016'][i] = temp_list[87]
        normalized_df['05-2016'][i] = temp_list[88]
        normalized_df['06-2016'][i] = temp_list[89]
        normalized_df['07-2016'][i] = temp_list[90]
        normalized_df['08-2016'][i] = temp_list[91]
        normalized_df['09-2016'][i] = temp_list[92]
        normalized_df['10-2016'][i] = temp_list[93]
        normalized_df['11-2016'][i] = temp_list[94]
        normalized_df['12-2016'][i] = temp_list[95]
    
    return normalized_df
```
